```python
#  import needed library
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#load data
import pandas as pd
data= pd.read_csv("../input/data11/Admission_Predict.csv")
data.info()
```

```python
data.head()
```

```python
#Dropping unwanted column
data = data.drop('gre', axis='columns')
data.head()
```

```python
#Dropping unwanted column
data = data.drop('sop', axis='columns')
data.head()
```

```python
#rename columns
data.columns=['GPA','Chance']
data.head()
```

```python
#check for missing value 
data.isnull().any()
```

no missing value

```python
#visualizing the data
plt.scatter(data.GPA,data.Chance, marker='+', color = 'red')
```

```python
data.shape
```

```python
#split the data for trainig and testing 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(data[['GPA']],data.Chance,test_size=0.35)
```

```python
#modeling 
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
```

```python
#train model
model.fit(X_train,y_train)
```

```python
#test model
model.predict(X_test)
```

```python
# check accuracy 
model.score(X_test,y_test)
```

```python
# visualizing fit 
X_new = np.linspace(6, 11, 100).reshape(-1, 1)
y_proba = model.predict_proba(X_new)
plt.scatter(data.GPA,data.Chance, marker='+', color = 'red')
plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2 ,label='GPA')
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2 ,label='Chance')
```
